```python
from ultralytics import YOLO
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from IPython.display import Image,display
from IPython import display
```

```python
import cv2

from ultralytics import solutions

cap = cv2.VideoCapture("Path/to/video/file.mp4")
assert cap.isOpened(), "Error reading video file"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

# Video writer
video_writer = cv2.VideoWriter("distance_calculationss.avi", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

# Init distance-calculation obj
distance = solutions.DistanceCalculation(model="yolo11n.pt", show=True)

# Process video
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or video processing has been successfully completed.")
        break
    im0 = distance.calculate(im0)
    video_writer.write(im0)
    cv2.imshow('Distance Calculation', im0)
    if cv2.waitKey(1) == ord('q'):  # Press 'q' to quit
        break

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
from ultralytics.solutions import distance_calculation
import cv2

# Load model
model = YOLO("../models/yolov8n.pt")

# Initialize video
cap = cv2.VideoCapture("../test_vids/people_Walking.mp4")
assert cap.isOpened(), "Error opening video file"

# Init distance calculation object
dist_calc = distance_calculation.DistanceCalculation()

while cap.isOpened():
    success, frame = cap.read()
    if not success: break

    # Perform tracking
    tracks = model.track(frame, persist=True, show=False)
    
    # Calculate and draw distances
    frame = dist_calc.start_process(frame, tracks)
    
    # Optional: Display the frame
    cv2.imshow('Frame', frame)
    if cv2.waitKey(1) == ord('q'): break

cap.release()
cv2.destroyAllWindows()
```
